```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs
```

```python
df = pd.read_csv("ds/hour.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
sbs.barplot(data = df, x='hr', y='cnt')
```

```python
df.columns
```

```python
for c in ['season', 'yr', 'mnth', 'hr']:
    sbs.barplot(data=df, x=c, y='cnt')
    plt.show()
```

```python
sbs.pairplot(data=df[['temp', 'hum', 'windspeed', 'cnt']])
```

```python
sbs.jointplot(data=df, x='temp', y='cnt', kind='kde')
```

```python
sbs.barplot(data=df, x='hr', y='cnt', hue='workingday')
```

```python
sbs.relplot(data=df, x='hr', y='cnt', hue='season', col='workingday', kind='line')
```
