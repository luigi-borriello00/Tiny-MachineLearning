```python
def plot_letter(X, title):
    for row in X:
        xs, ys = [], []
        x, y = 0, 0
        ax = row[0::3]
        ay = row[1::3]
        xs.append(x)
        ys.append(y)
        for el in ax:
            x = x + el
            xs.append(x)
        for el in ay:
            y += el
            ys.append(y)
        plt.title(title)
        plt.plot(xs, ys)
```
